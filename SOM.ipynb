```python
from minisom import MiniSom

from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dim = 25
digits = datasets.load_digits(n_class=5)
data = digits.data
som = MiniSom(dim, dim, 64, sigma=.8, learning_rate=0.5)
som.train_random(data, 2500)
```

```python
classes = set(digits.target)
target = digits.target
neurons = np.full((dim, dim), np.nan)
for i, x in enumerate(data):
    r = som.winner(x)
    neurons[r] = target[i]
notnan_index = np.where(np.isfinite(neurons))
notnan_neurons = neurons[notnan_index]
plt.figure(dpi=300)
plt.hexbin(notnan_index[0], notnan_index[1], notnan_neurons, gridsize=dim, cmap='Paired')
plt.show()
```
